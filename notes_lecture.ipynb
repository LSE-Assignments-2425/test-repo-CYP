#### Example of random.randn()

```python
import numpy.random as nr
import matplotlib.pyplot as plt

prng = nr.randn(1000000)
print(prng)
plt.hist(prng, bins=2000)
plt.show()
```

#### sys example

```python
import sys

print(sys.path)
print(sys.version)
```

```python
import sys
import math
# Try: excecuted codes
# Except = reports the error when try session is not executable
# If no exception, execute else block
# Finally, always do this, whenever error occurs of not

try: 
    print(math.sqrt(-1))
except: 
    print(sys.exc_info())
else: 
    print("No error!")
finally:
    print("Finally finished")

```

### 6. Numpy visualizing

```python
import matplotlib.pyplot as plt
import numpy as np

points = np.arange(-5, 5, 0.01)
# print(points)
xs, ys = np.meshgrid(points, points)
# print(xs)
# print(ys)
# print(np.meshgrid(points,points))
z = np.sqrt(xs**2 + ys**2)
print(np.size(z))
plt.imshow(z, cmap="hot")
plt.colorbar()
plt.title("Image of $\sqrt{x^2+y^2}$ for a grid of values")
plt.show()
```
